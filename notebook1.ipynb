```python
missing_questions = {} 
```

```python
orgs = data.DEMO_Org.unique()
```

```python
for org in orgs:
    missing_questions[org] = data[data.DEMO_Org == org].isna().sum()
```

```python
new_data = pd.DataFrame(columns = data.columns.insert(0,'Organisation'))
```

```python
i = 0; 
for org in orgs:
        l = missing_questions[org].tolist()
        l.insert(0,org)
        new_data.loc[i] = l
        i += 1 
```
