```python
## Creating a dictionary of all rate questions that need to be aligned to scale 

rates = {
    "1": ["OBSV_PreC_interative_A_inperson" , "OBSV_PreC_interative_B_remote"], 
    "2": ["OBSV_PreC_meetingsize0ffice_A_1",
"OBSV_PreC_meetingsize0ffice_B_2n3",
"OBSV_PreC_meetingsize0ffice_C_4n9",
"OBSV_PreC_meetingsize0ffice_D_10plus"], 
    "3":["OBSV_PreC_meeting_A_inperson",
    "OBSV_PreC_meeting_B_hyrbid",
    "OBSV_PreC_meeting_C_remote"] ,
    "4":[
        "OBSV_same_space_count_A_1",
"OBSV_same_space_count_B_2n3",
"OBSV_same_space_count_C_4n9",
        "OBSV_same_space_count_D_10p"
    ], 
    "5":[
        "OBSV_PreC_Location_A_primryoffice",
"OBSV_PreC_Location_B_officeother",
"OBSV_PreC_Location_C_client",
"OBSV_PreC_Location_D_WFH",
"OBSV_PreC_Location_E_publicspace",
"OBSV_PreC_Location_F_transit",
"OBSV_PreC_Location_G_other"
    ], 
    "6":[
        "OBSV_PreC_wiOffice_A_desk",
"OBSV_PreC_wiOffice_B_otherdeck",
"OBSV_PreC_wiOffice_C_meetingrm",
"OBSV_PreC_wiOffice_D_lobby",
"OBSV_PreC_wiOffice_E_speicaltyarea",
"OBSV_PreC_wiOffice_F_other"
    ], 
    "7":["OBSV_Now_Interact_Work_within_org",
"OBSV_Now_WrkTimeA_Indiv_fract_vCollab"]
}
```

```python
## Reading original datafile provided by plastarc
data = pd.read_csv("WPA_DataCase_2024.csv") 
```

```python
## Fillling False NAs with True Zero Values 

for key in rates.keys():
    data[rates[key]] = data[rates[key]].fillna(0)
    
pd.set_option('display.max_columns', None) 

pd.set_option('display.max_rows', None)
```

```python
## Checking if sum for a particular "rate question" equals one or not 
## If not then normalizing 


for i in range (len(data)):
    for rate in rates.values(): 
        if np.sum(data.loc[i,rate].values) == 0: 
            data.loc[i,rate] = data.loc[i,rate].replace({0:np.nan}) 
        elif np.sum(data.loc[i,rate].values) == 1: 
            continue 
        else:
            a = np.sum(data.loc[i,rate].values) 
            for j in range(len(rate)-1):
                data.loc[i,rate[j]] = data.loc[i,rate[j]]/a  
            data.loc[i,rate[-1]] = 1 - np.sum(data.loc[i,rate[:-1]]) 
    print(i) 
```

```python
## Sanity Check loop 

for i in range(len(data)): 
    for rate in rates.values():
        print(np.sum(data.loc[i,rate].values))
```

```python
## Label Encoding Categorical Variables 

update1 = {
    "Very unimportant":1, 
    "Unimportant":2, 
    "Neutral":3, 
    "Important":4, 
    "Very important":5
} 
update2 = {
    'Very unsatisfied':1,
    'Very dissatisfied':1,
    'Unsatisfied':2, 
    'Dissatisfied':2,
    'Neutral':3, 
    'Satisfied':4, 
    'Very satisfied':5
} 
update3 = {
    'Strongly disagree':1, 
    'Disagree':2, 
    'Neutral':3, 
    'Agree':4, 
    'Strongly agree':5 
} 
```

```python
## Sheet names for in house datasheet containing combined variables 
sheets = ["Org1_","Org 2","Org 3","Org 4","Org 5"]

u = pd.read_excel("DataDictionary.xlsx", sheet_name=sheets[0]) 
```

```python
## Different combinations for different preference questions corresponding to above updates 


updates = {
    "1" : ['Recommendations Questions','Work-Life Balance Decision Factors','Community & Support Decision Factors',
      'Technology Decision Factors','Office Environment Decision Factors','Workplace Features Importance','Work Satisf Features', 
      ] ,
    "2":['Office Satisfaction','Office Satisfaction 1','Home Satisfaction'], 
    "3": ['Remote Work Indicators']
}
```

```python
## Label Encoding Loop 

for i in range(len(sheets)):
    u = pd.read_excel("DataDictionary.xlsx", sheet_name=sheets[i])  
    for j in range(1,4):
        for t in updates[str(j)]: 
            if i == 0 or i == 2:
                cols = u[u['Variable Name - Updated'] == t]['Variable Name'] 
            elif i == 1 or i == 3:
                cols = u[u['Variable Name - Updated'] == t]['Variable name '] 
            else:
                cols = u[u['Variable Name - Updated'] == t]['Variable name'] 
            for col in cols: 
                if j == 1:
                    data = data.replace({col:update1})
                elif j == 2:
                    data = data.replace({col:update2})
                else:
                    data = data.replace({col:update3})
```
